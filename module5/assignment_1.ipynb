#### 1. Write an essay: Importance of Data Cleaning in Data Science

Data cleaning is the foundational process of rectifying, formatting, and removing erroneous, duplicate, or incomplete data within a dataset, crucial for ensuring accuracy, reliability, and validity in data science

In the field of data science, there is a fundamental principle that guides every project: "Garbage In, Garbage Out". No matter how sophisticated a machine learning algorithm or how advanced a neural network may be, its output is only as reliable as the data used to train it. Data cleaning, the systematic process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset, is therefore not just a preliminary step—it is the foundation upon which all trustworthy insights are built.

### 1. The dataset contains more than 10,000 rows and more than 10 columns whichcontains features of the car and its (MSRP) manufacturer's suggested retail price. Clean the data and analyse it making it ready for modelling.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df.head()
```

#### Check the missing values

```python
df.isnull().sum()
```

```python
df['Market Category'] = df['Market Category'].fillna('Unknown')
```

```python
df.info()
```

##### Drop the null values of engine fuel type, engine hp, engine cylinders, numbers of doors 

```python
df.dropna(subset=['Engine Fuel Type','Engine HP','Engine Cylinders','Number of Doors'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

### EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_columns = df.select_dtypes(exclude=object)
```

```python
categorical_columns = df.select_dtypes(object)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(numerical_columns.corr(),cmap='coolwarm',annot=True)
plt.show()
```

```python
top_10 = df['Make'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_10.plot(kind='bar')
plt.title("Top 10 car brands")
plt.ylabel('Number of selling')
plt.show()
```

```python
df.head(5)
```

```python
top_make = df['Make'].value_counts().head(10).index
```

```python
top_make
```

```python
avg_price = df[df['Make'].isin(top_make)].groupby('Make')['MSRP'].mean()
```

```python
avg_price
```

```python
avg_price.sort_values().plot(kind='barh', figsize=(8,5))
plt.title("Average MSRP of Top 10 Brands")
plt.show()
```

```python
df['Car Age'] = 2026 - df['Year']

df.groupby('Car Age')['MSRP'].mean().plot()
plt.title("Price vs Car Age")
plt.show()
```
