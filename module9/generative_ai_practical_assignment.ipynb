## 2. Zomato-Style Restaurant Info Chatbot

This is a simple rule-based chatbot. It takes a restaurant name and cuisine as input and returns a canned response.


```python
restaurants = {
    'Pizza Palace': 'Italian',
    'Spice Hub': 'Indian',
    'Sushi World': 'Japanese',
    'Burger Barn': 'American',
    'Tandoori Flames': 'North Indian'
}

def get_restaurant_info(name, cuisine):
    name = name.strip().title()
    cuisine = cuisine.strip().title()

    if name in restaurants and restaurants[name].lower() == cuisine.lower():
        return f'Yes, {name} serves {cuisine} cuisine.'
    elif name in restaurants:
        return f'No, {name} serves {restaurants[name]} cuisine, not {cuisine} cuisine.'
    else:
        return f'Sorry, I do not have information about {name}.'

# Example outputs
print(get_restaurant_info('Pizza Palace', 'Italian'))
print(get_restaurant_info('Burger Barn', 'Indian'))
print(get_restaurant_info('Sushi World', 'Japanese'))
```

## 4. Spotify-Style Song Filter Function

### Prompt Used with ChatGPT / GitHub Copilot

Write a Python function that takes a list of song names and returns only those songs that start with the letter **S**. The check should be case-insensitive.

### Code Generated


```python
def filter_songs_starting_with_s(songs):
    return [song for song in songs if song.lower().startswith('s')]

# Example playlist
playlist = [
    'Shape of You',
    'Sunflower',
    'Believer',
    'Satranga',
    'Perfect',
    'See You Again',
    'Thunder',
    'Senorita'
]

filtered_playlist = filter_songs_starting_with_s(playlist)
print(filtered_playlist)
```

## 5. Production Challenges in Generative AI Chatbots

| Challenge | Example Scenario | Suggested Solution |
|---|---|---|
| Hallucinations / inaccurate answers | A BookMyShow chatbot may give the wrong movie timing, ticket price, or theatre availability. | Use Retrieval-Augmented Generation (RAG) so the chatbot fetches real-time data from the official database before answering. |
| High cost and latency at scale | A Myntra chatbot may receive thousands of user queries during a sale, making AI responses slow and expensive. | Use smaller optimized models for simple queries, caching for repeated answers, and route only complex queries to larger models. |

### Extra Note

Generative AI chatbots are powerful, but in production they must be connected with real databases, monitoring systems, safety filters, and fallback logic. This helps companies avoid wrong answers, slow responses, and poor user experience.
